```python
from bs4.element import Tag


import requests
from openai import OpenAI
from bs4 import BeautifulSoup

OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')

# Fetching the webpage from the url

def fetch_webpage(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # removing the junk tags

    for tag in soup(['script', 'style', 'header']):
        tag.decompose()

    return soup.get_text(strip=True)[:5000]


def summarize_webpage(url):
    response = ollama.chat.completions.create(
        model="llama3.2:1b",
        messages=[
            {'role': 'system', 'content': f"Summarize the following webpage: {fetch_webpage(url)}"},
            {'role': 'user', 'content': f"Summarize the content in 3 bullet points"}
            ]
    )
    return response.choices[0].message.content




url = "https://en.wikipedia.org/wiki/Large_language_model"

content = fetch_webpage(url)

response = requests.get(url)

summary = summarize_webpage(url)

print(summary)

```
